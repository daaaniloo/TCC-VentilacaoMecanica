```python

mapa = {1:'Sim', 2:'Não', 9:'Ignorado',
        '1':'Sim', '2':'Não', '9':'Ignorado',
       1.0:'Sim', 2.0:'Não', 9.0:'Ignorado',
       }

mapa_suport = {1:'Invasivo', 2:'NãoInvasivo', 3:'Não'} 

cols1 = ['CS_GESTANT', 
         'PUERPERA',   
         'CARDIOPATI',
         'HEMATOLOGI',
         'SIND_DOWN',
         'HEPATICA',
         'ASMA',
         'DIABETES',
         'NEUROLOGIC',
         'PNEUMOPATI',
         'IMUNODEPRE',
         'RENAL',
         'OBESIDADE',   
        'NOSOCOMIAL',
        'FATOR_RISC',
         'FEBRE',
        'TOSSE',
        'GARGANTA',
        'DISPNEIA',
        'DESC_RESP',
        'SATURACAO',
        'DIARREIA',
        'VOMITO',
        'DOR_ABD',
        'FADIGA',
        'PERD_OLFT',
        'PERD_PALA', 
        'VACINA',
        'ANTIVIRAL',
        'SUPORT_VEN',
        'VACINA_COV']
```

```python
for cols in cols1:
  if cols == 'SUPORT_VEN':
    df[cols]=df[cols].map(mapa_suport)
  else:
    df[cols]=df[cols].map(mapa)
    print(cols)
    print(df[cols].unique())
    print("------------------")
    print('\n')
```
