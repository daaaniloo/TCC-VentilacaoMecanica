```python
df_dummies = df.drop(['SUPORT_VEN','FATOR_RISC'], axis=1)
```

```python
lista_carac_2 = []
lista_carac_mais = []

for col in df_dummies.columns:
  if len(df_dummies[col].unique()) > 2:
    lista_carac_mais.append(col)
  else:
    lista_carac_2.append(col)
```

```python
df_dummies = pd.get_dummies(df_dummies[lista_carac_mais])
```

```python
for col in lista_carac_2:
  df_dummies[col] = df[col]
```

```python
df_dummies['SUPORT_VEN'] = df['SUPORT_VEN'].map({'Invasivo': 1, 'NãoInvasivo': 2, 'Não': 3})
```
